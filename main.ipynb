```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# Configurar opções de exibição do pandas
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.width', 1000)
```

```python
# leitura dos dados

categories = pd.read_csv('data/categories.csv', delimiter=';') 
customers = pd.read_csv('data/customers.csv', delimiter=';')
employee_territories = pd.read_csv('data/employee_territories.csv', delimiter=';') 
employees = pd.read_csv('data/employees.csv', delimiter=';')
order_details = pd.read_csv('data/order_details.csv', delimiter=';') 
orders = pd.read_csv('data/orders.csv', delimiter=';') 
products = pd.read_csv('data/products.csv', delimiter=';') 
region = pd.read_csv('data/region.csv', delimiter=';')
shippers = pd.read_csv('data/shippers.csv', delimiter=';') 
suppliers = pd.read_csv('data/suppliers.csv', delimiter=';') 
territories = pd.read_csv('data/territories.csv', delimiter=';')
us_states =  pd.read_csv('data/us_states.csv', delimiter=';')
```

```python
# mostrar tabelas para analisar o que pode ser necessário  na limpeza e transformação dos dados
from IPython.display import display

# Exibir tabelas com os 5 primeiros
print("Categories")
display(categories.head())
print("\nCustomers")
display(customers.head())
print("\nEmployee Territories")
display(employee_territories.head())
print("\nEmployees")
display(employees.head())
print("\nOrder Details")
display(order_details.head())
print("\nOrders")
display(orders.head())
print("\nProducts")
display(products.head())
print("\nRegion")
display(region.head())
print("\nShippers")
display(shippers.head())
print("\nSuppliers")
display(suppliers.head())
print("\nTerritories")
display(territories.head())
print("\nUS States")
display(us_states.head())
```

```python
#Limpeza e Transformação dos Dados

# 1. Orders
# 2. Order_Details
# 3. Products
```

```python
# Transformação de tipos de dados
orders['order_date'] = pd.to_datetime(orders['order_date'])
orders['required_date'] = pd.to_datetime(orders['required_date'])
orders['shipped_date'] = pd.to_datetime(orders['shipped_date'])

order_details['unit_price'] = order_details['unit_price'].astype(float)
order_details['quantity'] = order_details['quantity'].astype(int)
order_details['discount'] = order_details['discount'].astype(float)

products['unit_price'] = products['unit_price'].astype(float)
products['units_in_stock'] = products['units_in_stock'].astype(int)
products['units_on_order'] = products['units_on_order'].astype(int)
products['reorder_level'] = products['reorder_level'].astype(int)
products['discontinued'] = products['discontinued'].astype(bool)
```

```python
# Verificar e tratar valores nulos
print("Valores nulos em 'orders':")
print(orders.isnull().sum())
print("\nValores nulos em 'order_details':")
print(order_details.isnull().sum())
print("\nValores nulos em 'products':")
print(products.isnull().sum())
```

```python
# Tratamento de valores nulos
orders['ship_region'] = orders['ship_region'].fillna('Unknown')
orders['ship_postal_code'] = orders['ship_postal_code'].fillna('Unknown')
```

```python
# Verificação de duplicatas
print("Duplicatas em 'orders':")
print(orders.duplicated().sum())
print("\nDuplicatas em 'order_details':")
print(order_details.duplicated().sum())
print("\nDuplicatas em 'products':")
print(products.duplicated().sum())
```

```python
# Unir tabelas
order_products = pd.merge(order_details, products, on='product_id')
order_products_orders = pd.merge(order_products, orders, on='order_id')

display(order_products_orders.head())
```

```python
# Calcular faturamento total

order_products_orders['total_price'] = order_products_orders['quantity'] * order_products_orders['unit_price_x'] #cria coluna com total_price

faturamento_total = order_products_orders['total_price'].sum()
print(f'Faturamento Total: {faturamento_total}')

# Faturamento por categoria
faturamento_por_categoria = order_products_orders.groupby('category_id')['total_price'].sum().reset_index()
faturamento_por_categoria = pd.merge(faturamento_por_categoria, categories, on='category_id')
print(faturamento_por_categoria)

# Agrupar por região e somar o faturamento
faturamento_por_regiao = order_products_orders.groupby('ship_region')['total_price'].sum().reset_index()
print(faturamento_por_regiao)
```

```python
# Ticket Médio por Cliente
ticket_medio_por_cliente = order_products_orders.groupby('customer_id')['total_price'].mean().reset_index()
ticket_medio_por_cliente.columns = ['customer_id', 'ticket_medio']
print('ticket_medio_por_cliente')
print(ticket_medio_por_cliente)

# Vendas por Funcionario 
vendas_por_funcionario = order_products_orders.groupby('employee_id')['total_price'].sum().reset_index()
vendas_por_funcionario.columns = ['employee_id', 'vendas_por_funcionario']
print('vendas_por_funcionario')
print(vendas_por_funcionario)
```

```python
# Taxa de Churn -  calculada com base nos clientes que não fizeram pedidos nos últimos meses

ultimo_pedido = orders.groupby('customer_id')['order_date'].max().reset_index() 

#criada coluna churn que recebe o valor 1 se a data do último pedido for mais antiga 
#do que a data atual menos seis meses, indicando que o cliente não fez pedidos nos últimos seis meses. 
ultimo_pedido['churn'] = (ultimo_pedido['order_date'] < (orders['order_date'].max() - pd.DateOffset(months=6))).astype(int)
 
taxa_churn = ultimo_pedido['churn'].mean() #media coluna churn 
print(f'Taxa de Churn: {taxa_churn}')
```

```python
# visualizando dados


# Salvar os gráficos como imagens
plt.figure(figsize=(10, 6))
sns.barplot(x='category_name', y='total_price', data=faturamento_por_categoria)
plt.title('Faturamento por Categoria de Produto')
plt.xlabel('Categoria')
plt.ylabel('Faturamento')
plt.xticks(rotation=45)
plt.savefig('faturamento_por_categoria.png')  # Salvar o gráfico como uma imagem

plt.figure(figsize=(10, 6))
sns.barplot(x='ship_region', y='total_price', data=faturamento_por_regiao)
plt.title('Faturamento por Região')
plt.xlabel('Região')
plt.ylabel('Faturamento')
plt.xticks(rotation=45)
plt.savefig('faturamento_por_regiao.png')  # Salvar o gráfico como uma imagem

plt.figure(figsize=(10, 6))
sns.histplot(ticket_medio_por_cliente['ticket_medio'], bins=30)
plt.title('Distribuição do Ticket Médio por Cliente')
plt.xlabel('Ticket Médio')
plt.ylabel('Frequência')
plt.savefig('ticket_medio_por_cliente.png')  # Salvar o gráfico como uma imagem

plt.figure(figsize=(8, 8))
plt.pie(vendas_por_funcionario['vendas_por_funcionario'], labels=vendas_por_funcionario['employee_id'], autopct='%1.1f%%')
plt.title('Total de Vendas por Funcionário')
plt.savefig('vendas_por_funcionario.png')  # Salvar o gráfico como uma imagem

```
